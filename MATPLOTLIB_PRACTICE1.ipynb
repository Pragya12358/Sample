```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize=(5,3), dpi = 300)
plt.plot(x, y, 'r.--', label = '2x', markersize = 10, markeredgecolor = 'b')

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2,'g', label = 'x^2')
plt.plot(x2[5:], x2[5:]**2, 'g--')

plt.title("My first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0, 1, 2, 3])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show() 
```
